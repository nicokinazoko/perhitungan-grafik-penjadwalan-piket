```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dataMemetika = pd.read_csv('data-memetika.csv')

dataMemetika.dtypes

waktuProses = dataMemetika['Waktu Proses']
jumlahGenerasi = dataMemetika['Jumlah Generasi']
jumlahPopulasi = dataMemetika['Tanggal Jumlah Populasi']
crossoverRate = dataMemetika['Crossover Rate']
mutationRate = dataMemetika['Mutation Rate']
nilaiFitnessTiga = dataMemetika['Nilai Fitness 3']
nilaiFitnessDua = dataMemetika['Nilai Fitness 2']
nilaiFitnessSatu = dataMemetika['Nilai Fitness 1']
nilaiFitnessNol = dataMemetika['Nilai Fitness 0']
nilaiFitnessNegatifSatu = dataMemetika['Nilai Fitness -1']
nilaiFitnessKosong = dataMemetika['Nilai Fitness Kosong']
dataMemetika['Nilai Fitness Min'] = dataMemetika['Nilai Fitness -1'] + \
    dataMemetika['Nilai Fitness Kosong']
nilaiFitnessMin = dataMemetika['Nilai Fitness Min']
```

```python
dataSortingFitnessMin = dataMemetika.sort_values(by='Waktu Proses', ascending=False)
dataSortingFitnessMin.head()
```

```python
# sorting data berdasarkan populasi
dataSortingPopulasi = dataMemetika.sort_values(by='Tanggal Jumlah Populasi', ascending=False)
dataSortingPopulasi.head()
```

```python
# Hubungan antara Waktu Proses dengan Jumlah Nilai Fitness Terendah dan Kosong

plt.style.use('seaborn')
plt.title('Data Fitness Min dengan Runtime')
plt.xlabel('Waktu Proses (s)')
plt.ylabel('Jumlah Nilai Fitness Min')
plt.scatter(waktuProses, nilaiFitnessMin, c='blue',
            edgecolor='black', linewidth=1)
plt.tight_layout()
plt.show()
```
