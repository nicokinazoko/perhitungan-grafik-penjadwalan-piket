```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dataGenetika = pd.read_csv('data-genetika.csv')

dataGenetika.dtypes

waktuProses = dataGenetika['Waktu Proses']
jumlahGenerasi = dataGenetika['Jumlah Generasi']
jumlahPopulasi = dataGenetika['Tanggal Jumlah Populasi']
crossoverRate = dataGenetika['Crossover Rate']
mutationRate = dataGenetika['Mutation Rate']
nilaiFitnessTiga = dataGenetika['Nilai Fitness 3']
nilaiFitnessDua = dataGenetika['Nilai Fitness 2']
nilaiFitnessSatu = dataGenetika['Nilai Fitness 1']
nilaiFitnessNol = dataGenetika['Nilai Fitness 0']
nilaiFitnessNegatifSatu = dataGenetika['Nilai Fitness -1']
nilaiFitnessKosong = dataGenetika['Nilai Fitness Kosong']
dataGenetika['Nilai Fitness Min'] = dataGenetika['Nilai Fitness -1'] + \
    dataGenetika['Nilai Fitness Kosong']
nilaiFitnessMin = dataGenetika['Nilai Fitness Min']
```

```python
dataSortingFitnessMin = dataGenetika.sort_values(by='Nilai Fitness 3', ascending=False)
dataSortingFitnessMin.head()
```

```python
dataRerata = dataGenetika.mean(axis=0)
print(dataRerata)
```

```python
# Hubungan antara Waktu Proses dengan Jumlah Nilai Fitness Terendah dan Kosong

plt.style.use('seaborn')
plt.title('Data Fitness Min dengan Runtime')
plt.xlabel('Waktu Proses (s)')
plt.ylabel('Jumlah Nilai Fitness Min')
plt.scatter(waktuProses, nilaiFitnessMin, c='blue',
            edgecolor='black', linewidth=1)
plt.tight_layout()
plt.show()
```
